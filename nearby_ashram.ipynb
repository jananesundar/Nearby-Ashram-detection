```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap
import folium
from geopy.distance import geodesic
```

```python
df = pd.read_csv("C:/Users/Janane Sundar/Desktop/name.csv")
df
```

```python
X = df[['Name', 'Longitude', 'Latitude']]
X
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
kmeans.fit(X[['Latitude', 'Longitude']])
X['cluster_label'] = kmeans.labels_

# Getting input for the source data point
source_lat = float(input("Enter source latitude: "))
source_lon = float(input("Enter source longitude: "))
source_location = np.array([[source_lat, source_lon]])

# Test the source data point
source_cluster = kmeans.predict(source_location)

# Find nearby locations within the same cluster
nearby_locations = X[X['cluster_label'] == source_cluster[0]]

# Calculate geodesic distances and sort by distance
nearby_locations['distance'] = nearby_locations.apply(
    lambda row: geodesic((source_lat, source_lon), (row['Latitude'], row['Longitude'])).kilometers,
    axis=1
)
nearby_locations = nearby_locations.sort_values(by='distance').head(4)

# Create a folium map centered around the mean latitude and longitude
map_center = [X['Latitude'].mean(), X['Longitude'].mean()]
my_map = folium.Map(location=map_center, zoom_start=12)

# Add markers for each data point with different colors for each cluster
for index, row in X.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color=ListedColormap(['blue', 'green', 'orange'])(row['cluster_label']),
        fill=True,
        fill_color=ListedColormap(['blue', 'green', 'orange'])(row['cluster_label']),
        fill_opacity=0.7,
        popup=f"Ashram: {row['Name']}\nCluster: {row['cluster_label']}",
    ).add_to(my_map)

# Add a marker for the source data point with a blue color
folium.Marker(
    location=[source_lat, source_lon],
    popup=f"Source Location - Ashram: {X.loc[X['cluster_label'] == source_cluster[0], 'Name'].values[0]}\n"
          f"Cluster: {source_cluster[0]}",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(my_map)

# Add markers for the 4 nearby locations with red color and display distances
for index, row in nearby_locations.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Ashram: {row['Name']}\nCluster: {row['cluster_label']}\n"
              f"Geodesic Distance: {row['distance']:.2f} kilometers",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(my_map)

# Save the map to an HTML file
my_map.save("map1.html")

# Print the 4 nearby locations
print("4 Nearby Locations:")
for index, row in nearby_locations.iterrows():
    print(f"Ashram: {row['Name']},    Distance: {row['distance']:.2f} km")
```
